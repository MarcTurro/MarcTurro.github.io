## Impl√©mentation du labyrinthe

Impl√©mentons ce labyrinthe avec un graphe.
Vous utiliserez l'impl√©mentation de graphe fournie ci-dessous.  
Vous nommerez les sommet (1, 1) (1, 2) .. etc 

Les ar√™tes relient les cases voisines quand le passage est possible. Si une paroie emp√™che de passer on ne met pas l'ar√™te.    
Par exemple il n'y a pas d'ar√™te entre les sommets (1, 1) et (1, 2) mais il y en a une entre (1, 1) et (2, 1).

```python
class Graph() :
    def __init__(self) :
        self.lst_adj = {}
        
    def __str__(self):
        return str(self.lst_adj)
    
    def add_sommet(self, sommet):
        self.lst_adj[sommet]=[]
        
    def add_arete(self, a, b) :
        self.lst_adj[a].append(b)
        self.lst_adj[b].append(a)
```

```python
labyrinthe = Graph()
for i in range(1, 5):
    for j in range(1, 9):
        labyrinthe.add_sommet((i, j))
labyrinthe.add_arete((1, 1), (2, 1))
labyrinthe.add_arete((2, 1), (2, 2))
labyrinthe.add_arete((2, 2), (2, 3))
labyrinthe.add_arete((2, 2), (3, 2))
labyrinthe.add_arete((2, 3), (1, 3))
labyrinthe.add_arete((1, 3), (1, 4))
labyrinthe.add_arete((1, 4), (1, 5))
labyrinthe.add_arete((1, 4), (2, 4))
labyrinthe.add_arete((1, 5), (2, 5))
labyrinthe.add_arete((2, 4), (3, 4))
labyrinthe.add_arete((2, 5), (2, 6))
labyrinthe.add_arete((2, 6), (1, 6))
labyrinthe.add_arete((2, 6), (3, 6))
labyrinthe.add_arete((2, 7), (3, 7))
labyrinthe.add_arete((1, 6), (1, 7))
labyrinthe.add_arete((2, 6), (2, 7))
labyrinthe.add_arete((1, 7), (1, 8))
labyrinthe.add_arete((1, 8), (2, 8))
labyrinthe.add_arete((3, 2), (4, 2))
labyrinthe.add_arete((4, 2), (4, 3))
labyrinthe.add_arete((4, 3), (4, 4))
labyrinthe.add_arete((3, 4), (3, 5))
labyrinthe.add_arete((3, 5), (3, 6))
labyrinthe.add_arete((3, 6), (4, 6))
labyrinthe.add_arete((3, 6), (3, 7))
labyrinthe.add_arete((4, 6), (4, 5))
labyrinthe.add_arete((3, 7), (4, 7))
labyrinthe.add_arete((4, 7), (4, 8))
labyrinthe.add_arete((4, 8), (3, 8))
labyrinthe.add_arete((2, 8), (3, 8))
```

## Dessin de votre labyrinthe
Pour voir le labyrinthe que vous avez cr√©√©, vous allez avoir besoin des fonction ci-dessous.

Vous n'avez pas besoin de les regarder, la seule chose dont vous avez besoin c'est de dessiner avec `show_Labyrinthe(cote, nli, ncol)` avec :
* `cote` est un entier qui code la longuer des cot√© des case (mettez 40)
* `nli` et `ncol` sont les nombres de lignes et colones de votre labyrinthe, donc ici 4 et 8

```python
from turtle import *
pencolor('green')
width(5)

def show_Labyrinthe(labyrinthe, cote, nli, ncol):
    speed(10)
    cadre(nli, ncol, cote)
    paroies_verticales(nli, ncol, cote)
    paroies_horizontales(nli, ncol, cote)

def cadre(nli, ncol, cote):
    penup()
    pendown()   
    # dessin des paroies ext√©rieures
    for _ in range(2):
        for i in range(ncol):
            forward(cote)
        right(90)
        for i in range(nli):
            forward(cote)
        right(90)

def paroies_verticales(nli, ncol, dist):
    """dessin des paroies internes verticales"""
    for i in range(1, nli + 1):
        for j in range(1, ncol) :
            dess = (i, j + 1) not in labyrinthe.lst_adj[(i, j)]
            if dess :
                penup()
                forward(dist)
                pendown()
                right(90)
                forward(dist)
                backward(dist)
                left(90)
            else :
                penup()
                forward(dist)

        penup()
        goto(0, 0)
        right(90)
        forward(dist*i)
        left(90)

def paroies_horizontales(nli, ncol, cote):
    """dessin des paroies internes horizontales"""
    penup()
    goto(0, -cote)
    for i in range(1, nli):
        for j in range(1, ncol) :
            dess=(i + 1, j) not in labyrinthe.lst_adj[(i, j)]
            if dess :
                pendown()
                forward(cote)
            else :
                penup()
                forward(cote)

        penup()
        goto(0, -cote*(i + 1))
```

```python
show_Labyrinthe(labyrinthe, 40, 4, 8)
mainloop()
```

```python
def dot(color):
    """Trace les points de couleur color"""
    pendown()
    pencolor(color)
    width(10)
    forward(1)
    penup()
    backward(1)
    pencolor('green')  # On remet la tortue en vert
    width(5)  # On remet la tortue en √©paisseur 5


def show_Parcours(labyrinthe, vus, cote, nli, ncol):
    show_Labyrinthe(labyrinthe, 40, 4, 8)
    speed(6)

    for s in vus :
        penup()
        goto(int((s[1] - 0.5)*cote), -5 - int((s[0] - 0.5)*cote)) 
        pendown()
        dot('red')
       
```
